```python
#trimming the audio to 30 sec even if it was 30 minuite long 
import whisper
model = whisper.load_model("base")
#we pad and trim to fit 30sec exactly as what the model needs
audio=whisper.load_audio("E:/ai and data science/record anaylsis/data/The Devils often in the Detail by Ben Henderson (Part 1 of 4).wav")
audio=whisper.pad_or_trim(audio)

mel=whisper.log_mel_spectrogram(audio).to(model.device)
#detect spoken lanuage
_,probs=model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

#decade the audio
options = whisper.DecodingOptions(fp16=False)
result = whisper.decode(model, mel, options)
print(result.text)


```
